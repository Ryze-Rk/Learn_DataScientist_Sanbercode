<h2>Soal 2: Aritmatika di Python </h2>
    
- Buat statement pertambahan antara 2 number di Python
- Buat statement perkurangan antara 2 number di Python
- Buat statement perkalian antara 2 number di Python
- Buat statement pembagian antara 2 number di Python

```python
a = 10
b = 3
print (a+b) #operasi pertambahan
print (a-b) #operasi perkurangan
print (a*b) #operasi perkalian
print (a/b) #operasi pembagian 
print (a // b) #floor division, pembagian dengan menampilkan hasil pembagian yang merupakan nilai pembulatan kebawah
print (a % b) #Modulus, pembagian dengan menampilkan hasil sisa dari operasi
```

<h2>Quiz 3: Assign Variable dan Tipe Data Integer, Float </h2>

- Buat suatu variabel a dan b, dimana a dan b adalah nilai bertipe data numeric
- Berikan suatu nilai bertipe data integer, hasil pembagian dari a dengan b
- Berikan suatu nilai bertipe data float, hasil pembagian dari a dengan b

```python
a = 5
b = 2
print ('### nilai operasi pembagian merupakan tipe data integer ###')
#jika hasil pembagian seharusnya ada koma, ketika di assign menjadi tipe data integer akan dibulatkan kebawah hasilnya
print (int (a/b))
print(type(int (a/b)))
print ('### nilai operasi pembagian merupakan tipe data float ###')
print (float (a/b))
print(type(float (a/b)))
```
