<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    print ('# Pengecekan argumen #')
    print(inp_list)
    print(sum(inp_list))
    print(len(inp_list))
print(mean_list(obj_list)) #jawaban akan null/none karena tidak ada operasi return dari pembagian
```

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    return(sum(inp_list)/len(inp_list)) 
    #nama fungsi baru mean_list(inp_list) mengandung argumen obj_list 
    #yang perlu dilakukan adalah mendefinisikan sum (jumlah angka di objek list) dibagi dengan len (mengembalikan jumlah item dalam objek ) dan mengembalikannya sebelum di tampilkan ke output 
print(mean_list(obj_list))
```

Expected Output:

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(list1, list2):
    print ('# Pengecekan argumen #')
    print(list1)
    print(list2)
print(kali_list(obj_list, obj_penambah)) #jawaban akan null/none karena tidak ada operasi return dari penggabungan list
```

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(list1, list2):
    list1.extend(list2)
    return(list1)
# nama fungsi baru, kali_list(list1, list2) mengandung argumen obj_list sebagai list1 dan obj_penambah sebagai list2
# penggabungan antara 2 list biasanya menggunakan extend method dimana list2 akan ditambahkan ketika list1 di kembalikan untuk ditampilkan pada output
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
