- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'or'
- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'not'

```python
print( "=== contoh gabungan multiple comparison menggunkan Boolean Comparisin 'and'===" )
a = 11 > 7 # True
b = 2 < 5 # True
print((a) and (b)) 
c = 11 < 7 # nilai perbandingan variabel c adalah False karena 11 tidak mungkin lebih kecil dari 7
d = 2 < 5 # True
print((c) and (d)) 
# boolean comparisin 'and' hanya bernilai True jika dan hanya jika kedua perbandingan gabungan True
# sedangkan jika salah satu dari kedua perbandingan salah maka nilai comparisin gabungan keduanya akan False

print( "=== contoh gabungan multiple comparison menggunkan Boolean Comparisin 'or'===" )
e = 9 == 9 # True
f = 7 != 7 # nilai perbandingan variabel adalah False karena 7 seharusnya sama dengan 7 nilainya sama bukan tidak sama
print((e) or (f)) 
g = 9 != 9 # False
h = 11 != 11 # False
print((g) or (h)) 
i = 13 == 13 # True
j = 15 == 15 # True
print((i) or (j)) 
# boolean comparisin 'or' hanya bernilai True jika salah satu perbandingan True ataupun jika kedua perbandingan bernilai True
# sedangkan jika kedua perbandingan nilainya False maka nilai comparisin gabungan keduanya akan False

print( "=== contoh gabungan multiple comparison menggunkan Boolean Comparisin 'not'===" )
k = 3 > 1 # True
l = 1 != 1 # False
m = 3 == 3 #True
print(k is not l) 
print (k is not m)
# is not berarti membandingkan dua multiple comparison dan menilai True/False sesuai dengan letak awal masing-masing comparisonnya misal T is not F berarti True karena benar tidaklah salah
# tetapi jika T is not T berarti False karena seharusnya True is True adalah benar

# not simpelnya adalah kebalikan nilai dari comparison
print( "=== contoh Boolean Comparisin 'not'===" )
o = 3
l = 1
print (not(o > l)) # seharusnya 3 lebih dari 1 True, akan tetapi ketika diberi not berarti kebalikannya yaitu False





```

<h2>Soal 3: If-Else Statement</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

- Bualah sebuah if-else statement yang dimana akan mem-print 'Besar' jika ruangan adalah 'Kamar' dan ukuran lebih dari 12, kemudian mem-print 'Sedang' jika ruangan adalah 'Kamar' dan ukuran lebih dari 6 dan memprint 'Kecil' jika ruangan adalah 'Kamar' dan ukuran ruangan lebih kecil dan sama dengan 6.

```python
print ('=== PERBANDINGAN UKURAN KAMAR VERSI 1 ===')
ruangan = 'Kamar'
size = 7

if ruangan == 'Kamar' and size > 12:
    print('Besar')
elif ruangan == 'Kamar' and size > 6:
    print('Sedang')
else:
    print('Kecil')

print('=== PERBANDINGAN UKURAN KAMAR VERSI 2 ===')
ruangan = 'Kamar'
size_1 = 7

if ruangan == 'Kamar' and size_1 > 12:
    print('Besar')
if ruangan == 'Kamar' and size_1 > 6 :
    print('Sedang')
if ruangan == 'Kamar' and size_1 <= 6:
    print('Kecil')
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

```python
# While loop digunakan untuk mengeksekusi blok pernyataan berulang kali selama bernilai True hingga kondisi tertentu terpenuhi.
print('=== Contoh While Loop ===')
count = 0

count = 2
while count < 5:
   print (count, " < 5")
   count = count + 2
else:
   print (count, " >= 5, pengulangannya udahan dulu")

# For loop, melakukan pengulangan dengan iterasi yang sudah diketahui sampai iterasi itu berhenti baik sesuai dengan definisi awal atau sampai dengan kondisi tertentu.
# For loop dapat digunakan pada data urutan (sequence) seperti list, tuple, string, atau objek apapun yang bisa diiterasikan.
print('=== Contoh For Loop sederhana ===')
for var in list(range(5)):
   print (var)

print('=== Contoh For Loop pada list string ===')
questions = ['Kumaha kabarna gengs?', 'Aeh mang kumaha damang?', 'Sugan tiasa dongkap kadieu?']
for q in questions:
    print(q)
```

<h2>Soal 6: While Loop</h2>
    
Buatlah sebuah code while loop sebagai berikut :
- Buatlah suatu fungsi yang merima satu input nilai numeric
- Dalam fungsi, Buat sebuah while-loop dengan melakukan comparasi terhadap inputan tersebut, dimana looping/iterasi terus berjalan apabila nilai inputan tidak sama dengan 0
- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
a = 10

def fungsi_while(a): # Fungsi yang menerima satu input nilai numerik
    while a != 0:
        print(a)
        if a > 0 :
          a -= 1
        else : # Sebenarnya bisa juga menggunakan if a < 0 :, tetapi karena menurut soal "selain itu" berarti menggunakan else :
          a += 1

fungsi_while(a)
```
