<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 



*   **Series pada pandas mempunyai index atau key yang dapat disesuaikan** sedangkan **list hanya mempunyai index saja** dan **dictionary hanya mempunyai key saja**
*   **Dataframe merupakan sekumpulan dari  data series** dimana objeknya 2 dimensi, punya index baris dan kolom sedangkan **data series hanya terdiri dari single list dengan index** dimana objeknya 1 dimensi



```python
# contoh data series
kota_RegJabar = pd.Series(['Jakarta', 'Bandung', 'Bogor', 'Tasikmalaya', 'Garut', 'Majenang'], index=['kota_1', 'kota_2', 'kota_3', 'kota_4', 'kota_5', 'kota_6']) 
kota_RegJabar
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# deskriptif statistik
df_diamonds.describe()
```

```python
# nilai rata-rata dari column price
print(df_diamonds['price'].mean())

# nilai standar deviasi dari column depth
print(df_diamonds['depth'].std())

# nilai maximum dari column carat
print(df_diamonds['carat'].max())
```

Jawab:



*   Metode pandas untuk mengetahui statistik deskriptif dari suatu data adalah :

    "nama variabel".describe()

*   Nilai rata-rata dari column price: 3932.7998

* Nilai standar deviasi dari column depth: 1.4326

* Nilai maksimum dari column carat: 5.01
