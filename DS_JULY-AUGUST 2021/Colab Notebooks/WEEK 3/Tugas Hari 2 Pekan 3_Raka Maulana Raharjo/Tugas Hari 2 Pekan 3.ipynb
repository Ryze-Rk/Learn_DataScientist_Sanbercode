Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#implisit
plt.figure(figsize=(10,7))
plt.plot(x,cos, label='Cosx', color='r', linestyle='solid')
plt.plot(x,cos3, label='3 Cosx', color='g', linestyle='dashed')
plt.plot(x,sin, label='sin(x-90)', color='b', linestyle='-.')
plt.plot(x,sin2, label='2 Sinx', color='y', linestyle='None', marker='x')
plt.title('Grafik Cos Sin', font=())
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.legend(loc=10, fontsize='small')
plt.show()
```

```python
#eksplisit
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(x, cos, label='Cosx', color='red')
ax.plot(x, cos3, linestyle='--', label='3 Cosx', color='green')
ax.plot(x, sin, linestyle='-.', label='Sin(x-90)', color='blue')
ax.plot(x, sin2, linestyle='None', marker='x', label='2 Sinx', color='y')
ax.set_xlabel('X-Axis Label ')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
fig.legend( loc="center", fontsize='small')
plt.show()
```

"Lebih sesuai menggunakan metode **eksplisit** karena seharusnya legend mengikuti garis persamaannya begitu pula fontsize legend dibuat small."
