```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize =(17,8))
ax[0].plot(x,tan, color='b', label='tanX')
ax[1].plot(x,sin, color='r', label='sinX')
ax[2].plot(x,cos, color='g', label='cosX')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc=9,fontsize='medium')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(x,y)
ax.set_title('Grafik CosX')

ins1 = ax.inset_axes([0.9, 0.9, 0.3, 0.3]) #kanan atas
ins1.plot(x, y4, color='black')
ins1.plot(x,y,color='blue')
ins1.set_title('Grafik cosX^4')

ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3]) #kiri atas
ins2.plot(x, y5, color='y')
ins2.plot(x,y,color='blue')
ins2.set_title('Grafik cosX^5')

ins3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3]) #kanan bawah
ins3.plot(x, y3, color='green')
ins3.plot(x,y,color='blue')
ins3.set_title('Grafik cosX^3')

ins4 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3]) #kiri bawah
ins4.plot(x, y2, color='red')
ins4.plot(x,y,color='blue')
ins4.set_title('Grafik cosX^2')

plt.show()
```
