```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv') #import manual csv pakai file di google colab
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
group = df.groupby(['Genre'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
for i,j in enumerate(group.columns):
  print(i,j)
```

```python
import numpy as np
```

```python
grouptrans=group.transpose()
print(grouptrans)
```

```python
print("Jawaban soal 2.1")

plt.figure(figsize=(12,8))

x = np.arange(0, 16, 4)
width = 0.3
colors =['orangered','tab:blue','mediumpurple','gray','gold','yellowgreen','pink','orangered','tab:blue','mediumpurple','gray','gold']

for i,j in enumerate(grouptrans.columns):
  plt.bar(x+(i*width), grouptrans[j], width, label=j, color=colors[i])
 
plt.style.use('ggplot') #belum bisa berwarna legendny?
plt.xticks(x+1.6, labels=grouptrans.index)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games by Genre', fontweight='bold')
plt.legend(loc=(0.95,0.75), frameon=True)
plt.show()
```

Jawaban soal 2.2 :

**Dengan kuantitatif barplot kita dapat mengetahui perbandingan banyaknya data antar kategori dengan tinggi plot masing-masing, jika kuantitatif barplotnya semakin tinggi berarti datanya di suatu kategori  banyak dan jika semakin pendek berarti datanya di suatu kategori  sedikit.**

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
print("Jawaban soal 3.1")

plt.figure(figsize=(12,8))

x = grouptrans.index
width = 0.7
y_bottom = 0

colors =['orangered','tab:blue','mediumpurple','gray','gold','yellowgreen','pink','orangered','tab:blue','mediumpurple','gray','gold']

for i,j in enumerate(grouptrans.columns):
  plt.bar(x, grouptrans[j], width, label=j, color=colors[i], bottom=y_bottom)
  y_bottom += grouptrans[j]
 
plt.style.use('ggplot')
plt.legend(loc=(0.95,0.75), frameon=True)
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.title('Mean Sales Video Games by Genre', fontweight='bold')
plt.show()
```

Jawaban soal 3.2 : 

**Dengan menggunakan Stack Barplot kita dapat membandingkan total value pada beberapa kategori berdasarkan panjang data yang terlihat pada visualisasi, jika pada stack barplot semakin panjang maka berarti datanya di suatu kategori banyak dan jika semakin pendek berarti datanya di suatu kategori  sedikit.**
