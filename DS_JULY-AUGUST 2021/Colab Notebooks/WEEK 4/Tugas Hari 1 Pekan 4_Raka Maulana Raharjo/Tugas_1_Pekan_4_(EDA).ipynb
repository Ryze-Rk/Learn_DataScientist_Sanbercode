```python
dfvisual_nohandling = df_soal2.sample(1000)
dfvisual_nohandling
```

```python
pd.unique (dfvisual_nohandling['DewPointAvgF']) #cek apakah sudah terisi semua
```

```python
dfvisual_nohandling.isnull().sum()
```

```python
df_soal2.fillna(method="ffill", inplace=True)
```

```python
dfvisual_whandling = df_soal2.sample(1000)
```

```python
pd.unique (dfvisual_whandling['DewPointAvgF']) #cek apakah sudah terisi semua
```

```python
dfvisual_whandling.isnull().sum()
```

```python
print("Tanpa Handling Missing Value")
plt.figure(figsize=(8,4))
plt.scatter(x=dfvisual_nohandling['HumidityAvgPercent'],y=dfvisual_nohandling['DewPointAvgF'],c=dfvisual_nohandling['TempAvgF'],s=dfvisual_nohandling['WindAvgMPH']*20, cmap='coolwarm')
plt.xlabel('HumidityAvg (%)')
plt.ylabel('DewPointAvg (F)')
plt.title('Austin Weather')
plt.show()

print("Dengan Handling Missing Value")
plt.figure(figsize=(8,4))
plt.scatter(dfvisual_whandling['HumidityAvgPercent'], dfvisual_whandling['DewPointAvgF'],c=dfvisual_whandling['TempAvgF'],s=dfvisual_whandling['WindAvgMPH']*20, cmap='coolwarm')
plt.xlabel('HumidityAvg (%)')
plt.ylabel('DewPointAvg (F)')
plt.title('Austin Weather')
plt.show()
```

Analisa :

Jika pada awalnya data merupakan tipe object, data tersebut belum bisa divisualisasikan. Data yang bisa divisualisasikan adalah data berupa tipe numerik seperti float. Pada tugas kali ini, kita akan melakukan handling outliers. Outliers ditemukan pada kolom 'House_Price' dengan menggunakan boxplot dari library seaborn seperti gambar berikut:

Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah
*   Visualisasikan hasil data yang sudah d remove outliers seperti expected berikut:







**EXPECTED OUTPUT:**

Batas bawah & batas atas:

Visualisasi boxplot:

![boxplot2](https://drive.google.com/uc?id=1qZc-lEa5FUzSQ_cyt4_GJQ0TZ8c6WjHX)





```python
Q1=df2['House_Price'].quantile(0.25)
Q3=df2['House_Price'].quantile(0.75)
IQr=Q3-Q1 
lower_bound=Q1-IQr*1.5
upper_bound=Q3+IQr*1.5
```

```python
df_visualbox=df2[(df2['House_Price'] < upper_bound) & (df2['House_Price'] > lower_bound)]
df_visualbox
```

```python
print('Visualisasi Boxplot Sebelum di Interquartil dan di remove boxfliers:')
import seaborn as sns 
plt.figure(figsize=(12,8))
sns.boxplot(x='House_Price', data=df2)
plt.show()
```

```python
print('Batas bawah & batas atas:')
print(lower_bound,upper_bound)
print('Visualisasi Boxplot:')
plt.figure(figsize=(12,8))
sns.boxplot(x='House_Price',data=df_visualbox)
plt.show()
```
